Defining this method (command) will allow us to pause for some time between our code so that we can visualize what's going on.

````java
public static void await(double seconds)
{
    int ms = (int)(seconds * 1000);
    try {
        Thread.sleep(ms);
    } catch (Exception e) {
        System.out.println("Timeout?");
    }
}
````

What you've just done is print some text out to the **console**. The console is a text-based program that you can interact with. You give it commands, it runs them, or it says it doesn't recognize the command. When a Java program is run in a console, we can have it output some text to the user.

Let's try using the command above called `await` that we defined along with it.

````java
System.out.println("Waiting 3 seconds before outputting text...");
await(3);
System.out.println("Waited 3 seconds!");
await(2);
System.out.println("Waited 5 seconds now!");
````
